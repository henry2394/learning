Tuy nhin Implemantation ca Elasticnet trong Sklearn c tnh nh sau:



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
bikes  = pd.read_csv('hour.csv')
bikes.head()
bikes_prep = bikes.copy()
bikes_prep = bikes_prep.drop(['index','date','casual','registered'],axis = 1)
bikes_prep = bikes_prep.drop(['weekday', 'workingday','year','atemp','windspeed'], axis = 1)
bikes_prep.head()
```

```python
# Problems: Solve the problem of Autocorrelation, Thm cc ct lag ny vo


t_1 = bikes_prep['demand'].shift(+1).to_frame()
t_1.columns = ['t-1']

t_2 = bikes_prep['demand'].shift(+2).to_frame()
t_2.columns = ['t-2']

t_3 = bikes_prep['demand'].shift(+3).to_frame()
t_3.columns = ['t-3']

bikes_prep_lag = pd.concat([bikes_prep, t_1, t_2, t_3], axis=1)

# Xa dng b null 
bikes_prep_lag = bikes_prep_lag.dropna()

# Chuyn cc ct cn x l sang nh dng category

bikes_prep_lag['season'] = bikes_prep_lag['season'].astype('category')
bikes_prep_lag['holiday'] = bikes_prep_lag['holiday'].astype('category')
bikes_prep_lag['weather'] = bikes_prep_lag['weather'].astype('category')
bikes_prep_lag['month'] = bikes_prep_lag['month'].astype('category')
bikes_prep_lag['hour'] = bikes_prep_lag['hour'].astype('category')

# convert sang dng dummy, b ct u tin v k cn n cng c th bit c gi tr l g
bikes_prep_lag = pd.get_dummies(bikes_prep_lag, drop_first=True)
```

```python
Y = bikes_prep_lag[['demand']]
X = bikes_prep_lag.drop(['demand'], axis=1)

tr_size = 0.7 * len(X)
tr_size = int(tr_size)

X_train = X.values[0 : tr_size]
X_test = X.values[tr_size : len(X)]

Y_train = Y.values[0 : tr_size]
Y_test = Y.values[tr_size : len(Y)]
```

```python
from sklearn.linear_model import LinearRegression

std_reg = LinearRegression()
std_reg.fit(X_train, Y_train)

r2_train = std_reg.score(X_train, Y_train)
print("r2_train: ", r2_train)
r2_test  = std_reg.score(X_test, Y_test)
print("r2_test: ", r2_test)


# Create Y Predictions
Y_predict = std_reg.predict(X_test)


from sklearn.metrics import mean_squared_error
rmse = math.sqrt(mean_squared_error(Y_test, Y_predict))
print("rmse: ", rmse)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1)
ridge.fit(X_train, Y_train)
y_predict = ridge.predict(X_test)

print("r2_train: ", ridge.score(X_train, Y_train))

print("r2_test: ", ridge.score(X_test, Y_test))


rmse2 = math.sqrt(mean_squared_error(Y_test, y_predict))
print("rmse: ", rmse2)
```

```python
from sklearn.linear_model import Lasso
ls = Lasso(alpha=1)
ls.fit(X_train, Y_train)
y_predict = ls.predict(X_test)

print("r2_train: ", ls.score(X_train, Y_train))

print("r2_test: ", ls.score(X_test, Y_test))


rmse2 = math.sqrt(mean_squared_error(Y_test, y_predict))
print("rmse: ", rmse2)
```

```python
from sklearn.linear_model import ElasticNet

#LinerRegression
el = ElasticNet()
el.fit(X_train, Y_train)

y_predict = el.predict(X_test)

print("r2_train: ", el.score(X_train, Y_train))

print("r2_test: ", el.score(X_test, Y_test))


rmse2 = math.sqrt(mean_squared_error(Y_test, y_predict))
print("rmse: ", rmse2)
```

# Using Regularization for Feature Selecting

```python
from sklearn.linear_model import Lasso, Ridge, LinearRegression

#LinerRegression
lr = LinearRegression()
lr.fit(X,Y)
lr_coef = lr.coef_
lr_intercept = lr.intercept_

#Lasso
ls = Lasso(alpha=10)
ls.fit(X,Y)
ls_coef = ls.coef_
ls_intercept = ls.intercept_

#Ridge
rd = Ridge(alpha=10)
rd.fit(X,Y)
rd_coef = rd.coef_
rd_intercept = rd.intercept_
```
