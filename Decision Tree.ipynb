# I: Tổng quan

#####  Decision tree là một mô hình supervised learning, có thể được áp dụng vào cả hai bài toán classification và regression
   - Regression Tree có biến mục tiêu là biến liên tục, trong khi Classification Tree có biến mục tiêu la biến phân loại.
   - Trong Regression Tree, khi huấn luyện, giá trị tại nút lá bằng trung bình các giá trị biến mục tiêu của các điểm dữ liệu có trong nút đó. Nên khi đưa tập test vào, nếu các điểm dữ liệu rơi vào nút lá nào, kết quả trả ra sẽ là giá trị trung bình.
   - Với Classification Tree, khi huấn luyện, giá trị tại nút lá(phân lớp) bằng giá trị có tần suất cao nhất(Mode) của các dữ liệu trong nút đó. Nên khi đưa tập test vào, nếu các điểm dữ liệu rơi vào nút lá nào, kết quả trả ra sẽ là Mode.

   Tổng quan thì mô hình bắt đầu bằng việc chọn 1 feature là Root Nood, sau mỗi lần lặp của thuật toán sẽ sử dụng các feature chưa được dùng đến làm các node tiếp theo (Các kỹ thuật lựa chọn feature nào sẽ là node tiếp theo sẽ được trình bày chi tiết ở dưới). Thuật toán tiếp tục lặp lại trên mỗi tập con, chỉ lặp các feature chưa được chọn trước đó.

- Mô hình có thể dừng lại trong một trong những trường hợp sau:
      + Mọi output trong tập hợp con đều thuộc cùng một lớp
      + Không còn Input nào được chọn. Trong trường hợp Output trong tập con không thuộc cùng một lớp thì sẽ được gắn nhãn bằng output phổ biến nhất trong tập hợp con.

####  Bước 1: Chọn Root node

Bằng trực giác, một Root nood là tốt nhất nếu dữ liệu trong mỗi child node hoàn toàn thuộc vào một class–khi đó node này có thể được coi là một leaf node, tức ta không cần phân chia thêm nữa. 

Nếu dữ liệu trong các child node vẫn lẫn vào nhau, ta coi rằng phép phân chia đó chưa thực sự tốt. Từ đó ta cần có một hàm số đo độ tinh khiết (purity), hoặc độ vẩn đục (impurity) của một nodes

###### Đối với Classification Tree
Các phương pháp tính độ vẩn đục (so sánh khả năng giải thích dữ liệu): Gini Index, Information gain, Chi test, Reduce Variance --> Lựa chọn Feature làm Root node có giá trị min (tức là nó giải thích được nhiều dữ liệu hơn các feature khác)

###### Đối với Regression Tree
Tính Sum of Squared Residuals của Y tại nhiều điểm dữ liệu, chọn ra điểm có Residual thấp nhất làm mốc để chia dữ liệu. Trong trường hợp có nhiều feature thì tương tự như trên tuy nhiên ta sẽ tính thêm cho các feature còn lại, so sánh min(SSR) của các feature và chọn feature có giá trị min(SSR) thấp nhất làm root của tree

Giá trị ở mỗi Leaf được tính toán bằng average các giá trị trong nhóm đó

#### Bước 2: Phân nhánh

Tính toán Gini/Entrophy/SSR tại mỗi Nodes trước và sau khi phân nhánh nó để check xem liệu phân nhánh có tốt hơn không.

- TH1: Sau khi phân nhánh Gini/Entrophy/SSR < trước phân nhánh tại node đó ====> Tiếp tục phân nhánh

- TH2: Sau khi phân nhánh Gini/Entrophy/SSR > trước phân nhánh tại node đó ====> Dừng tại nodes đó

Tiếp tục làm như vậy đến khi đạt yêu cầu dừng ở trên hoặc việc phân nhóm là hoàn hảo (mỗi leave 1 class không bị lẫn)

####  Một số lưu ý khi làm việc với Feature

- Feature là dạng Yes/No làm bình thường
- Feature dạng Rank thì tính Gini/entrophy/SSr tại các rank, ví dụ: Rank <= 1, Rank <=2 ... 
- Feature dạng Multiple choice tính cho dừng choice và mọi possible combination. ví dụ: Feature color = ['red','blue'] --> Gini/Entrophy/SSr cho Red, Blue,Green, Red or Blue, Red or Green, Blue or Green

# II: Prunning Tree 
#### Mục tiêu : Tránh Overfit

#### Các tham số có thể tác động: 

1, criterion: Phương pháp chọn root node ( Gini hoặc entrophy)

2, splitter: Lựa chọn cách tách node, trong bộ tách “best” nó đánh giá tất cả các feature trong khi bộ tách “random” sử dụng một hàm thống nhất ngẫu nhiên với min_feature_value, max_feature_value và random_state làm đầu vào

3, Presort: Có sort data trước khi split không (Nhanh hơn và chính xác hơn)


4, Max_depth: Chiều tối đa của cây. Nếu để mặc định, thì các nút được mở rộng cho đến khi tất cả các lá đều thuần túy hoặc cho đến khi tất cả các lá chứa ít hơn mẫu min_samples_split. Max_dept càng cao thì thông tin nhận được sẽ nhiều hơn tuy nhiên dễ dẫn tới overfit


5, min_samples_split: Số lượng mẫu tối thiếu để tách node. Có thể nhập int hoặc float.
 - Nếu là int thì là số mẫu tối thiểu, 
 - Float thì sẽ là tỷ lệ, tức là số mẫu tối thiểu = float x Số lượng sample
 
 
6, min_samples_leaf: Leaf node là 1 node mà k có node con. Đây là số lượng phần tử tối thiểu yêu cầu để được coi là 1 leaf
 - Nếu là int thì là số mẫu tối thiểu, 
 - Float thì sẽ là tỷ lệ, tức là số mẫu tối thiểu = float x Số lượng sample
 
 
7, max_features: Số lượng các Features được xem xét khi tìm kiếm sự phân chia tốt nhất:
  - Nhập int: Số feature
  - Nhập float: max_features = float * n_features
  - Auto: thì max_features=sqrt(n_features).
  - Sqrt: max_features=sqrt(n_features).
  - log2: max_features=log2(n_features).
  - None: max_features=n_features.
  
  
8, Max leaf node (int, default=None): Số lượng lá tối đa


9, min_impurity_decrease (float, default=0.0): Một node sẽ được tách nếu sự phân tách này làm giảm impurity lớn hơn hoặc bằng giá trị này


10, min_impurity_split (float, default=0): Ngưỡng để dừng phân tách, 1 node se tách nếu impurity > giá trị này. còn nếu impurity bé hơn thì sẽ coi là 1 leaf


11, Class_weight: Thuật toán này thường hoạt động không tốt trên tập dữ liệu có nhãn k cân bằng. vì ở mỗi node khi bị tách, chúng bị chi phối bởi number of class. Nếu dữ liệu quá lệch về 1 class thì dataset là imbalance -> kq sẽ nghiêng về 1 class. DO vậy cần đánh trọng số để có kq tốt hơn. Class_weight = 'balance'




```python
import pandas as pd
data= pd.read_csv("04+-+decisiontreeAdultIncome.csv")
data.head()
```

```python
# Khong co missing value
data.isnull().sum()
data.dtypes
```

```python
#Create dummy
data_prep = pd.get_dummies(data, drop_first=True)
data_prep.head()
```

```python
# Define X va Y
X = data_prep.drop(['IncomeClass_ >50K'], axis=1)
Y = data_prep['IncomeClass_ >50K']

# Train/test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3, random_state = 1234, stratify = Y) 

# Model
from sklearn.tree import DecisionTreeClassifier
DC = DecisionTreeClassifier(criterion="gini", random_state=1234 )
DC.fit(x_train,y_train)
y_predict = DC.predict(x_test)

# Evaluate
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
score = DC.score(x_test, y_test)
print(cm)
print(score)
```

```python
from sklearn.tree import DecisionTreeClassifier
DC = DecisionTreeClassifier(criterion="entropy", random_state=1234, max_depth = 5)
DC.fit(x_train,y_train)
y_predict = DC.predict(x_test)

# Evaluate
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
score = DC.score(x_test, y_test)
print(cm)
print(score)
```
